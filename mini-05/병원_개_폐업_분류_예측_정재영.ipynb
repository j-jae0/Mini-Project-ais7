```python
def bedCount2band(num):
    if num < 30: return "의원"
    elif 30 <= num < 100: return "병원"
    elif 100 <= num < 500: return "종합병원"
    elif num >= 500: return "상급종합병원"
```

```python
df_train["bedCount"] = df_train["bedCount"].apply(bedCount2band)
df_test["bedCount"] = df_test["bedCount"].apply(bedCount2band)
```

## 전처리 : 결측치 대체

```python
# 결측치값을 포함하는 데이터를 모두 제거하기엔 데이터 크기가 너무 작음 -> -999로 대체
same_col = ["OC", "sido", "sgg", "bedCount", "instkind"]
in_col_train = df_train.columns.tolist()[6:]
in_col_test = df_test.columns.tolist()[6:]
```

```python
temp = df_train[in_col_train].replace(0, -999)
temp = temp.fillna(-999)

pre_train = pd.concat([df_train[same_col], temp], axis=1)
```

```python
temp = df_test[in_col_test].replace(0, -999)
temp = temp.fillna(-999)

pre_test = pd.concat([df_test[same_col], temp], axis=1)
```

```python
# ownerChange에서 nan은 유지
pre_train["ownerChange"] = pre_train["ownerChange"].replace(-999, np.nan)
pre_test["ownerChange"] = pre_test["ownerChange"].replace(-999, np.nan)
```

```python
display(pre_train.head())
pre_test.head()
```

```python
pre_train = pre_train.fillna(-999)
pre_test = pre_test.fillna(-999)
```

```python
pre_test.isnull().sum()
```

```python
print(pre_train.isnull().sum().sum())
pre_test.isnull().sum().sum()
```

## 전처리 : 범주형 -> 수치형 인코딩 작업

```python
obj_cols = ["OC", "sido", "bedCount", "instkind", "ownerChange"]
```

```python
# train
temp_arr = []
for col in obj_cols:
    temp_arr.append(pd.get_dummies(pre_train[col], drop_first=True))

temp = pd.concat(temp_arr, axis=1)
train = pd.concat([temp, pre_train.drop(columns=obj_cols, axis=1)], axis=1) 
```

```python
# test
temp_arr = []
for col in obj_cols:
    temp_arr.append(pd.get_dummies(pre_test[col], drop_first=True))

temp = pd.concat(temp_arr, axis=1)
test = pd.concat([temp, pre_test.drop(columns=obj_cols, axis=1)], axis=1) 
```

## 텐서플로우 모델 생성

```python
import tensorflow as tf
```

```python
# from sklearn.model_selection import train_test_split

# X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.15, stratify=y)
# print(f"X_train: {X_train.shape}\ny_train: {y_train.shape}\nX_valid: {X_valid.shape}\ny_valid: {y_valid.shape}")
```

```python
# 입력할 데이터 수 (컬럼)
input_shape = X.shape[1]
input_shape
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(128, input_shape=[input_shape]),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(1, activation='sigmoid')
])
```

## 모델 컴파일

```python
# 모델 컴파일
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# 모델 요약
model.summary()
```

```python
class PrintDot(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        if epoch % 100 == 0: print('')
        print('.', end='')

# val_loss 기준으로 값이 나아지지 않으면 멈추게 합니다.
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=100)
```

## 모델 학습

```python
# 학습하기
history = model.fit(X_train, y_train, epochs=1000, validation_split=0.15)
```

```python
# 학습결과의 history 값을 가져와서 비교하기 위해 데이터프레임으로 변환
df_hist = pd.DataFrame(history.history)
df_hist.tail()
```
